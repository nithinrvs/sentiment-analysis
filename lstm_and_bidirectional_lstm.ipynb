# **Importing Corpus**



```python
def read_file_sections(filename):
    sections = []
    current_section = ""

    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            if line.strip() == "__________________________":
                if current_section:
                    sections.append(current_section.strip())
                    current_section = ""
            else:
                current_section += line

    if current_section:
        sections.append(current_section.strip())

    return sections
```

```python
product_pos= read_file_sections(r"/content/product_pos.txt")
product_neg = read_file_sections(r"/content/product_neg.txt")
```

```python
book_pos = read_file_sections(r"/content/book_pos.txt")
book_neg = read_file_sections(r"/content/book_neg.txt")
```

```python
df_product_neg = pd.DataFrame({'text': product_neg, 'label': 0})
df_product_pos = pd.DataFrame({'text': product_pos, 'label': 1})
df_book_neg = pd.DataFrame({'text': book_neg, 'label': 0})
df_book_pos = pd.DataFrame({'text': book_pos, 'label': 1})

# Concatenate and shuffle
df = pd.concat([df_product_neg, df_product_pos,df_book_neg,df_book_pos], ignore_index=True)
#df = pd.concat([df_product_neg, df_product_pos], ignore_index=True)

df = shuffle(df).reset_index(drop=True)
```

# **Removing Punctuations**

```python
import pandas as pd
import string

# Define a function to remove punctuation
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

# Apply the function to the "text" column
df['text'] = df['text'].apply(remove_punctuation)
```

```python
from indicnlp.tokenize import indic_tokenize

# Function to tokenize text
def tokenize_text(text):
    return indic_tokenize.trivial_tokenize(text)

# Create new column 'tokenized_text' with tokenized text
df['tokenized_text'] = df['text'].apply(tokenize_text)

# Print the updated dataframe with tokenized text column
print(df)
```

```python
# Calculate max sequence length for tokenized text
max_seq_length = max(len(tokens) for tokens in df['tokenized_text'])
print("Maximum Sequence Length for Tokenized Text:", max_seq_length)
```

# **Padding**

```python
import pandas as pd
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Pad sequences to max_seq_length
max_seq_length = 259
padded_sequences = pad_sequences(df['tokenized_text'], maxlen=max_seq_length, dtype=object, padding='post', truncating='post', value='')

# Update dataframe with padded sequences
df['padded_text'] = padded_sequences.tolist()

# Print the updated dataframe with padded text column
print(df)
```

# **Getting Embeddings**

```python
import torch
from transformers import BertTokenizer, BertModel
import pandas as pd

# Load your dataframe 'df'
# Assuming 'tokenized_text' and 'padded_text' columns are already populated

# Load the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased')

# Define a function to get embeddings
def get_embeddings(text):
    tokens = tokenizer.encode(text, add_special_tokens=True, max_length=259, truncation=True)
    input_ids = torch.tensor(tokens).unsqueeze(0)  # Add batch dimension
    with torch.no_grad():
        outputs = model(input_ids)
        embeddings = outputs[0][:, 0, :]  # Extract the [CLS] token embedding
    return embeddings.tolist()

# Apply the function to your dataframe and update the 'text_embeddings' column
df['text_embeddings'] = df['padded_text'].apply(get_embeddings)

# Save or use the updated dataframe with embeddings
df.to_csv('df_with_embeddings.csv', index=False)
```

# **Bi-Directional LSTM**

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

# Assuming 'df' is your dataframe containing text and label columns
# Tokenize the text data to get the vocabulary size
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['text'])  # Assuming 'text' is the column containing your text data
num_words = len(tokenizer.word_index) + 1  # Add 1 for the padding token

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(df['text'])

# Pad sequences to ensure uniform length
max_seq_length = 259  # Maximum sequence length based on your requirements
X_padded = pad_sequences(sequences, maxlen=max_seq_length, padding='post')

# Define LSTM model with Bidirectional layer and additional layers
model = Sequential()
model.add(Embedding(input_dim=num_words, output_dim=128, input_length=max_seq_length))
model.add(Bidirectional(LSTM(units=128, return_sequences=True)))
model.add(Dropout(0.5))  # Adding dropout for regularization
model.add(Bidirectional(LSTM(units=64)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Assuming 'y' contains your labels
y = df['label']

# Train the model on the entire dataset
model.fit(X_padded, y, epochs=10, batch_size=32, validation_split=0.2)  # Assuming you want to use 20% of data for validation
```
